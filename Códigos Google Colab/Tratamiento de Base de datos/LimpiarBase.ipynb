```python
import pandas as pd
# Paso 1: Cargar la base de datos
data = pd.read_csv('resultado.csv')

# Limpiar la columna de etiquetas
data['Activity'] = data['Activity'].apply(lambda x: ' '.join(x.split()))  # Reemplaza m√∫ltiples espacios por uno
data['Activity'] = data['Activity'].str.strip()  # Elimina espacios al inicio y al final

# Si quieres eliminar duplicados en la columna de etiquetas
data['Activity'] = data['Activity'].apply(lambda x: ','.join(sorted(set(x.split(',')))))

# Guardar la base de datos limpia
data.to_csv('resultado_limpio.csv', index=False)

```
