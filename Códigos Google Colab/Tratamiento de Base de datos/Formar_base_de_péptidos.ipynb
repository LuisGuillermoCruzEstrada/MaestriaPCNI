```python
import os
import csv

# Ruta de la carpeta
folder_path = 'Base3/'
output_csv = 'resultado_final_50.csv'

# Lista todos los archivos en la carpeta con extensión .txt
text_files = [file_name for file_name in os.listdir(folder_path) if file_name.endswith('.txt')]

# Inicializa una lista para almacenar los datos
all_data = []
column_names = None

# Recorre cada archivo .txt
for file_name in text_files:
    file_path = os.path.join(folder_path, file_name)

    with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
        # Lee todas las líneas del archivo
        lines = file.readlines()

        # Usa la primera línea como nombre de columnas si aún no se ha definido
        if column_names is None:
            column_names = lines[0].strip().split('\t')

        # Lee las líneas restantes como filas de datos
        for line in lines[1:]:
            row = line.strip().split('\t')
            all_data.append(row)

# Escribir los datos en un archivo CSV de salida
with open(output_csv, 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    # Escribe los nombres de las columnas
    writer.writerow(column_names)
    # Escribe cada fila de datos
    writer.writerows(all_data)

print(f"Datos convertidos y guardados en {output_csv}")
```
