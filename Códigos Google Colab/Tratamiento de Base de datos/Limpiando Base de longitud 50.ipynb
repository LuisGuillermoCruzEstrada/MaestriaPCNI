```python
import pandas as pd

# Cargar el archivo CSV (reemplaza 'tu_archivo.csv' con el nombre de tu archivo)
df = pd.read_csv('resultado_final_50.csv')

# Seleccionar las columnas necesarias
df_filtered = df[['Title', 'Seqence', 'Length', 'Activity']].copy()

# Modificar la columna 'Activity' para que solo contenga 'Antibacterial' o 'Otros'
df_filtered['Activity'] = df_filtered['Activity'].apply(lambda x: 'Antibacterial' if 'Antibacterial' in str(x) else 'Otros')

# Guardar el nuevo DataFrame en un archivo CSV
df_filtered.to_csv('nueva_base.csv', index=False)

print("Archivo 'nueva_base.csv' creado con éxito.")
```
